```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("./DATASET_COPIA.csv", sep=";", encoding="utf-8" )
```

```python
display(dataset)
```

```python
agrupado = dataset.groupby('Publisher')['Sales'].sum().reset_index()
```

```python
dtA = dataset[(dataset['Release date'] >= 1990) & (dataset['Release date'] < 2000)]
dtB = dataset[(dataset['Release date'] >= 2000) & (dataset['Release date'] < 2010)]
dtC = dataset[(dataset['Release date'] >= 2010) & (dataset['Release date'] < 2020)]
dtD = dataset[(dataset['Release date'] >= 2020)]
```

```python
display(dtA)
```

```python
agrupadoA = dtA.groupby('Publisher')['Sales'].sum().reset_index()
agrupadoB = dtB.groupby('Publisher')['Sales'].sum().reset_index()
agrupadoC = dtC.groupby('Publisher')['Sales'].sum().reset_index()
agrupadoD = dtD.groupby('Publisher')['Sales'].sum().reset_index()
agrupadoFinal = dataset.groupby('Publisher')['Sales'].sum().reset_index()
```

```python
porcentagem_de_vendasA = pd.merge(agrupadoA, agrupadoFinal, on='Publisher', how='inner')
porcentagem_de_vendasA['divisao'] = porcentagem_de_vendasA['Sales_x'] / porcentagem_de_vendasA['Sales_y'] * 100

porcentagem_de_vendasB = pd.merge(agrupadoB, agrupadoFinal, on='Publisher', how='inner')
porcentagem_de_vendasB['divisao'] = porcentagem_de_vendasB['Sales_x'] / porcentagem_de_vendasB['Sales_y'] * 100

porcentagem_de_vendasC = pd.merge(agrupadoB, agrupadoFinal, on='Publisher', how='inner')
porcentagem_de_vendasC['divisao'] = porcentagem_de_vendasC['Sales_x'] / porcentagem_de_vendasC['Sales_y'] * 100

porcentagem_de_vendasD = pd.merge(agrupadoD, agrupadoFinal, on='Publisher', how='inner')
porcentagem_de_vendasD['divisao'] = porcentagem_de_vendasD['Sales_x'] / porcentagem_de_vendasD['Sales_y'] * 100
```

```python
display(porcentagem_de_vendasA)
display(porcentagem_de_vendasB)
display(porcentagem_de_vendasC)
display(porcentagem_de_vendasD)
```
